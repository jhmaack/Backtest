```python
import pandas as pd
import pickle
```

```python
## CRYPTO


tmp = pd.read_excel('data/crypto.xlsx',header=None).dropna()
symbols_crypto = list(tmp[0])

with open('data/symbols_crypto', 'wb') as f:
    pickle.dump(symbols_crypto,f)
    f.close()

```

```python
## STOCKS

tmp2 = pd.read_csv('data/nasdaq.csv')
symbols_stocks = list(tmp2['Symbol'])

with open('data/symbols_stocks', 'wb') as f:
    # f.write(str(symbols_stocks))
    pickle.dump(symbols_stocks,f)
    f.close()
```
